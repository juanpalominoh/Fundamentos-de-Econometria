## 2. Variable Categórica: Nominal y Ordenadas


```r
# Variable Nominal
animales_vector <- c("Elefante", "Jirafa", "Mono", "Caballo")
factor_animales_vector <- factor(animales_vector)
factor_animales_vector
```
