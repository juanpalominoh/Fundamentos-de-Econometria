La información de fallecidos por COVID-19 proviene del siguiente link: https://www.datosabiertos.gob.pe/dataset/fallecidos-por-covid-19-ministerio-de-salud-minsa

```r
fallecidos <- import("https://files.minsa.gob.pe/s/t9AFqRbXw3F55Ho/download")
fallecidos 
```

```r
fallecidos_peru <- fallecidos %>% 
  mutate(FECHA=as.character(FECHA_FALLECIMIENTO),
         año=substring(FECHA, first=1, last=4),
         mes=substring(FECHA, first=5, last=6),
         dia=substring(FECHA, first=7, last=8),
         Fecha=as.Date(paste(año,mes,dia,sep="-"),"%Y-%m-%d")) %>%
  group_by(Fecha) %>% 
  summarize(fallecidos=n())

fallecidos_peru
```

La información de casos positivos proviene del siguiente link: https://www.datosabiertos.gob.pe/dataset/casos-positivos-por-covid-19-ministerio-de-salud-minsa

```r
positivos <- import("https://files.minsa.gob.pe/s/eRqxR35ZCxrzNgr/download")
positivos
```

```r
positivos_peru <- positivos %>% 
  mutate(FECHA=as.character(FECHA_RESULTADO),
         año=substring(FECHA, first=1, last=4),
         mes=substring(FECHA, first=5, last=6),
         dia=substring(FECHA, first=7, last=8),
         Fecha=as.Date(paste(año,mes,dia,sep="-"),"%Y-%m-%d")) %>%
  group_by(Fecha) %>% 
  summarize(positivos=n()) 

positivos_peru
```

```r
positivos_peru <- positivos_peru %>% 
        drop_na(Fecha)
```

```r
names(positivos_peru)
names(fallecidos_peru)
```

```r
base_covid <- positivos_peru %>%
    full_join(fallecidos_peru, by=c("Fecha"))
base_covid
```

```r
# Reemplazar con ceros los missing values (NA)
base_covid %>% 
    mutate(positivos = replace_na(positivos, 0),
           fallecidos= replace_na(fallecidos,0)) %>%
    arrange(Fecha)
```

# **Datos de Tipo de Cambio**

La información se puede visualizar en el siguiente link: 
https://estadisticas.bcrp.gob.pe/estadisticas/series/diarias/tipo-de-cambio

```r
# Descargamos la página web
download_html(url ="https://estadisticas.bcrp.gob.pe/estadisticas/series/diarias/resultados/PD04637PD/html",
              file="tipo_cambio.html")
```

```r
# Leemos el archivo descargado
archivo_tc <- read_html("/content/tipo_cambio.html", 
                                     encoding = "UTF-8")
```

Extraer la información con html_nodes y html_text

```r
tipo_cambio <- archivo_tc %>% 
  html_nodes(".dato") %>% 
  html_text()

tipo_cambio <- as.numeric(substring(tipo_cambio, first=13, last=18))
tipo_cambio
```

```r
fecha <- archivo_tc %>% 
  html_nodes(".periodo") %>% 
  html_text()

fecha <- substring(str_remove(fecha, pattern="\n"), first=13, last=19)
fecha
```

```r
# Extraer año
year <- substring(fecha, first=6, last=7)
year <- as.Date(year, "%y")
year <- substring(year, first=1, last=4)
```

```r
# Extraer mes
mes <- substring(fecha, first=3, last=5)
```

```r
meses <- unique(mes)
```

```r
# Extraer dia
dia <- substring(fecha, first=1, last=2)
```

```r
df_tc <- tibble(
  Fecha = as.Date(paste(year,match(mes, meses), dia, sep="-"), format="%Y-%m-%d"),
  tc  = tipo_cambio
  )
df_tc
```

Unir base de covid con el tipo de cambio

```r
# Reemplazar con ceros los missing values (NA)
base_covid_tc <- base_covid %>%
    mutate(positivos = replace_na(positivos, 0),
           fallecidos= replace_na(fallecidos,0)) %>%
    inner_join(df_tc, by=c("Fecha")) %>%
    mutate(letalidad=fallecidos*100/positivos,
           letalidad=ifelse(letalidad=="Inf", 0, letalidad)) %>%
    arrange(Fecha) %>%
    mutate(ID = row_number()) %>%
    drop_na(tc)
base_covid_tc
```

# 1. Modelo MCO


```r
install.packages("sandwich")
library("sandwich")
```

```r
formula <- tc ~ letalidad
ols <- lm(formula, data=base_covid_tc)
summary(ols)
```

```r
resids <- ols$residuals
```

```r
data_frame_white <- as.data.frame(cbind(base_covid_tc, resids))
```

# 2. Pruebas de Heterocedasticidad


## 2.1. Test Durbin Watson


```r
install.packages("lmtest")
library(lmtest)
install.packages("car")
library(car)
```

```r
dwtest(ols,alternative = "two.sided",iterations = 1000)
```

```r
durbinWatsonTest(ols,simulate = TRUE,reps = 1000) 
# Se rechaza la H0 de no autocorrelación
```

## 2.2 Test Breusch-Godfrey


```r
bgtest(ols, order=2)
```

# 3. MCO con errores estándar Newey-West


```r
NW_VCOV <- NeweyWest(ols, 
                     lag =1, prewhite = F, 
                     adjust = T)

coeftest(ols, vcov=NW_VCOV)
# Más detalles en: https://www.econometrics-with-r.org/15-4-hac-standard-errors.html
```
